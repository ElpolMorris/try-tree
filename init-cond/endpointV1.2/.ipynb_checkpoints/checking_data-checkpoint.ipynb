```python
# statesUSA
# hvdUSA
# countiesUSA

statesUSA   = pd.read_csv("statesUSA.csv")
hvdUSA      = pd.read_csv("hvdUSA.csv")
countiesUSA = pd.read_csv("countiesUSA.csv")
```

```python
print(df_states_US.equals(statesUSA))
print(df_hvd_US.equals(hvdUSA))
print(df_counties_US.equals(countiesUSA))
```

```python
statesUSA = statesUSA.rename(columns = {'State_postal':'state_postal'})
hvdUSA    = hvdUSA.rename(columns = {'State_Postal':'state_postal'})

_ = pd.merge(hvdUSA, statesUSA, how = 'outer', on = ['DateTime', 'State', 'state_postal', 'FIPS_state'],\
                      left_on = None, right_on = None, left_index = False, right_index = False, sort = True)

seirhvdUSA = _.dropna()

for attr in seirhvdUSA.columns:
    if seirhvdUSA[attr].dtypes == "float64":
        seirhvdUSA[attr] = seirhvdUSA[attr].astype('int64')
```

```python
# df_states_US.to_csv('statesUSA.csv', index = False, encoding = 'utf-8')
# df_hvd_US.to_csv('hvdUSA.csv', index = False, encoding = 'utf-8')
# df_counties_US.to_csv('countiesUSA.csv', index = False, encoding = 'utf-8')
seirhvdUSA.to_csv('seirhvdUSA.csv', index = False, encoding = 'utf-8')
```
